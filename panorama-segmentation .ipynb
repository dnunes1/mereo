```python
import numpy as np
from low.panorama import panorama
import cv2
from skimage.transform import resize
from skimage.segmentation import felzenszwalb as graphseg
from skimage.segmentation import mark_boundaries
```

```python
def imread(imagefile,color=1):
    img = cv2.imread(imagefile,color)
    return cv2.resize(img,(400,300)) 
```

```python
image1 = imread('/Users/danilonunes/workspace/boo/boomage/image/boo-beer.jpg',0)
image2 = imread('/Users/danilonunes/workspace/boo/boomage/image/boo-beer2.jpg',0)
print image1.shape, image2.shape
```

```python
img1 = imread('/Users/danilonunes/workspace/boo/boomage/image/boo-beer.jpg')
img2 = imread('/Users/danilonunes/workspace/boo/boomage/image/boo-beer2.jpg')
print img1.shape, img2.shape
```

```python
detector = cv2.SURF(400)
kp1 = detector.detect(image1)
kp2 = detector.detect(image2)
```

```python
descriptor = cv2.DescriptorExtractor_create("SURF")
k1, d1 = descriptor.compute(image1, kp1)  # keypoints object
k2, d2 = descriptor.compute(image2, kp2) 
```

```python
matcher = cv2.DescriptorMatcher_create("FlannBased")
matches = matcher.knnMatch(d1, d2, k=2)
```

```python
def filter_matches(kp1, kp2, matches, ratio=0.75):
    mkp1 = []
    for m in matches:
        if len(m) == 2 and m[0].distance < m[1].distance * ratio:
            mkp1.append(m[0])
    return mkp1
good_matches = filter_matches(k1, k2, matches, ratio=0.75)
```

```python
obj = np.array([k1[m.queryIdx].pt for m in good_matches])
scn = np.array([k2[m.trainIdx].pt for m in good_matches])

homography_matrix = cv2.findHomography(obj, scn, cv2.cv.CV_RANSAC)
r1, c1 = image1.shape
r2, c2 = image2.shape
result = cv2.warpPerspective(img1, homography_matrix[0], (c1 + c2, r1))
result[:r2, :c2, :] = img2
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pylab
b,g,r = cv2.split(result)
pylab.rcParams['figure.figsize'] = (6.0, 7.0)
plt.imshow(cv2.merge([r,g,b]))
```

```python
g1 = graphseg(img1, scale=100, sigma=0.5, min_size=50)
g2 = graphseg(img2, scale=100, sigma=0.5, min_size=50)
graph = np.hstack((g1,g2))
imgs = np.hstack((img1,img2))
view = mark_boundaries(imgs, graph)
b,g,r = cv2.split(view)
pylab.rcParams['figure.figsize'] = (6.0, 7.0)
plt.imshow(cv2.merge([r,g,b]))
```

```python
# loop over the unique segment values
for (i, segVal) in enumerate(np.unique(g2[4,:])):
    # construct a mask for the segment
    print "[x] inspecting segment %d" % (i)
    mask = np.zeros(image2.shape[:2], dtype = "uint8")
    mask[g2 == segVal] = 255
 
# plt.imshow(cv2.bitwise_and(img1, img1, mask = mask))
d1 = cv2.bitwise_and(image2, image2, mask = mask)
plt.imshow(d1,cmap='gray')
```
